```python
import os
import subprocess 
import sys
os.chdir('../')
import numpy as np
import pandas as pd
```

```python
test_folders=['akt1','ampc','cp3a4','cxcr4','gcr','hivpr','hivrt','kif11']
parent_dir='/data/pharmnn_rl/dude/all/'
```

Generate baseline pharmacophores

```python
for folder in test_folders:
    receptor_file=parent_dir+'/'+folder+'/receptor_nowat.pdb'
    ligand_file=parent_dir+'/'+folder+'/crystal_ligand.mol2'
    os.makedirs(parent_dir+'/'+folder+'/queries/',exist_ok=True)
    json_file=parent_dir+'/'+folder+'/queries/pharmit.json' 
    os.system(f'./pharmit pharma -receptor {receptor_file} -in {ligand_file} -out {json_file}')
    os.system(f'python genqueries.py {json_file} {receptor_file}')
```

```python
f1_scores={}
df_ligand_feats=pd.DataFrame(columns=['system','min_feats','model','precision','recall','f1','enrichment','hit_rate','num_feats'])
for folder in test_folders:
    print(folder)
    folder_dir=parent_dir+'/'+folder+'/'
    f1_scores[folder]=[]
    for file in os.listdir(parent_dir+'/'+folder+'/queries/'):
        if file !='pharmit.json':
            pharmacophore_file=folder_dir+'queries/'+file
            output=subprocess.check_output('python getf1.py '+pharmacophore_file+ ' ' +folder_dir+'pharmit_ligand_db' +' --actives ' + folder_dir +'actives_final.ism --decoys '+folder_dir+'decoys_final.ism' ,shell=True)
            output=output.decode('utf-8')
            results=output.split()
            f1=float(results[1])
            recall=float(results[3])
            precision=float(results[5])
            hit_rate=float(results[8])
            enrichment=float(results[11])
            num_feats=int(results[14])
            gh=float(results[16])
            df_ligand_feats=df_ligand_feats.append({'system':folder,'min_feats':3,'model':'random','precision':precision,'recall':recall,'f1':f1,'enrichment':enrichment,'hit_rate':hit_rate,'Guner-Henry':gh,'num_feats':num_feats},ignore_index=True)
            
        
            
```

```python
df_ligand_feats.to_csv('pharmit_ligand_random_results.csv')
```

```python
df_ligand_feats=pd.read_csv('pharmit_ligand_random_results.csv',index_col=0)
```

Obtain pharmacophores for each of the test receptors

```python
for test_folder in test_folders:
    print(test_folder)
    folder_dir=parent_dir+test_folder+'/'
    # ligand-only min 3 and min 4
    os.system('python predict.py --receptor '+folder_dir+'/receptor_nowat.pdb --ligand '+folder_dir+'/crystal_ligand.mol2 --features ligand_only --output_prefix '+folder_dir+'/ligand_min_3 --min_features 3')    
    os.system('python predict.py --receptor '+folder_dir+'/receptor_nowat.pdb --ligand '+folder_dir+'/crystal_ligand.mol2 --features ligand_only --output_prefix '+folder_dir+'/ligand_min_4 --min_features 4')
    #cnn min 3 and min 4
    os.system('python predict.py --receptor '+folder_dir+'/receptor_nowat.pdb --ligand '+folder_dir+'/crystal_ligand.mol2 --features cnn_only --output_prefix '+folder_dir+'/cnn_min_3 --min_features 3')
    os.system('python predict.py --receptor '+folder_dir+'/receptor_nowat.pdb --ligand '+folder_dir+'/crystal_ligand.mol2 --features cnn_only --output_prefix '+folder_dir+'/cnn_min_4 --min_features 4')
```

Get f1, precision,recall, hit rate and enrichment for each of those pharmacophores

```python
import numpy as np
import pandas as pd
import pickle
```

```python
df=pd.DataFrame(columns=['system','min_feats','model','precision','recall','f1','enrichment','hit_rate','num_feats'])
min_feats=[3,4]
prefixes=['ligand_min_','cnn_min_']
for test_folder in test_folders:
    print(test_folder)
    folder_dir=parent_dir+test_folder+'/'
    for prefix in prefixes:
        model_list=['model_ligand']
        if 'cnn' in prefix:
            model_list=['model_cnn_1','model_cnn_2','model_cnn_3','model_cnn_4','model_cnn_5']
        for model in model_list:
            for feat in min_feats:
                pharmacophore_file=folder_dir+prefix+str(feat)+'_'+model+'_predicted_pharmacophores.json'
                output=subprocess.check_output('python getf1.py '+pharmacophore_file+ ' ' +folder_dir+'pharmit_ligand_db' +' --actives ' + folder_dir +'actives_final.ism --decoys '+folder_dir+'decoys_final.ism' ,shell=True)
                output=output.decode('utf-8')
                results=output.split()
                f1=float(results[1])
                recall=float(results[3])
                precision=float(results[5])
                hit_rate=float(results[8])
                enrichment=float(results[11])
                num_feats=int(results[14])
                gh=float(results[16])
                df=df.append({'system':test_folder,'min_feats':feat,'model':model,'precision':precision,'recall':recall,'f1':f1,'enrichment':enrichment,'hit_rate':hit_rate,'Guner-Henry':gh,'num_feats':num_feats},ignore_index=True)
print(df.head())
df.to_csv('dude_prediction_results.csv')
```

Obtain mean and max pharmacophore from ligand scores

```python
ligand_means_f1={}
ligand_best_f1={}
for test_folder in test_folders:
    ligand_means_f1[test_folder]=df_ligand_feats[(df_ligand_feats['system']==test_folder)]['f1'].mean()
    ligand_best_f1[test_folder]=df_ligand_feats[(df_ligand_feats['system']==test_folder)]['f1'].max()
    print(test_folder,ligand_means_f1[test_folder],ligand_best_f1[test_folder])
```

```python
for test_folder in test_folders:
    df=df.append(df_ligand_feats[df_ligand_feats['system']==test_folder][df_ligand_feats['f1']==ligand_best_f1[test_folder]])
```

df to latex

```python
df.set_index(['system','model','min_feats'],inplace=True)
print(df.to_latex(index=True,sparsify=True,float_format="%.3f",multirow=False))
```

normalize f1 scores by the best

```python
df=pd.read_csv('dude_prediction_results.csv',index_col=0)
print(df)

for test_folder in test_folders:
     df=pd.concat([df,pd.DataFrame({'system':test_folder,'num_feats':"3-5",'model':'ligand_mean','f1':ligand_means_f1[test_folder]},index=[0])],ignore_index=True)
     df=pd.concat([df,pd.DataFrame({'system':test_folder,'num_feats':"3-5",'model':'ligand_best','f1':ligand_best_f1[test_folder]},index=[0])],ignore_index=True)
print(df)
```

```python
for test_folder in test_folders:
    df.loc[df.system == test_folder,'f1']/=ligand_best_f1[test_folder]
    ligand_means_f1[test_folder]/=ligand_best_f1[test_folder]
```

```python
matplotlib.rcParams.update({'font.size': 12})
ax=sns.boxplot(x='system',y='f1',data=df[df['model']=='PharmRL_CNN'],hue='model',whis=[0,100])
sns.stripplot(x='system',y='f1',data=df[df['model']=='PharmRL_Ligand_3'],ax=ax,hue='model',linewidth=1,edgecolor='black',palette=['orange'])
sns.stripplot(x='system',y='f1',data=df[df['model']=='PharmRL_Ligand_4'],ax=ax,hue='model',linewidth=1,edgecolor='black',palette=['red'])
sns.stripplot(x='system',y='f1',data=df[df['model']=='Ligand_mean'],ax=ax,hue='model',linewidth=1,edgecolor='black',palette=['green'])
handles, labels = ax.get_legend_handles_labels()
for ha in handles:
    ha.set_edgecolor('black')
    ha.set_linewidth(1)
lgd=ax.legend(handles, labels, loc='upper right')
ax.hlines(1,-1,8,linestyles='dashed',colors='black')
ax.set_ylabel('Ligand normalized F1 score')
ax.set_xlabel('Target')
plt.xticks(rotation=45)
plt.savefig('dude_f1score_results.pdf', bbox_inches='tight')
```
