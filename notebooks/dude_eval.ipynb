```python
import os
import subprocess 
import sys
os.chdir('../')
```

```python
test_folders=['akt1','ampc','cp3a4','cxcr4','gcr','hivpr','hivrt','kif11']
parent_dir='/data/pharmnn_rl/dude/all/'
```

Obtain pharmacophores for each of the test receptors

```python



for test_folder in test_folders:
    print(test_folder)
    folder_dir=parent_dir+test_folder+'/'
    # ligand-only min 3 and min 4
    os.system('python predict.py --receptor '+folder_dir+'/receptor_nowat.pdb --ligand '+folder_dir+'/crystal_ligand.mol2 --features ligand_only --output_prefix '+folder_dir+'/ligand_min_3 --min_features 3')    
    os.system('python predict.py --receptor '+folder_dir+'/receptor_nowat.pdb --ligand '+folder_dir+'/crystal_ligand.mol2 --features ligand_only --output_prefix '+folder_dir+'/ligand_min_4 --min_features 4')
    #cnn min 3 and min 4
    os.system('python predict.py --receptor '+folder_dir+'/receptor_nowat.pdb --ligand '+folder_dir+'/crystal_ligand.mol2 --features cnn_only --output_prefix '+folder_dir+'/cnn_min_3 --min_features 3')
    os.system('python predict.py --receptor '+folder_dir+'/receptor_nowat.pdb --ligand '+folder_dir+'/crystal_ligand.mol2 --features cnn_only --output_prefix '+folder_dir+'/cnn_min_4 --min_features 4')
```

Get f1, precision,recall, hit rate and enrichment for each of those pharmacophores

```python
import numpy as np
import pandas as pd
import pickle
```

```python
df=pd.DataFrame(columns=['system','num_feats','model','precision','recall','f1','enrichment','hit_rate'])
num_feats=[3,4]
prefixes=['ligand_min_','cnn_min_']
for test_folder in test_folders:
    print(test_folder)
    folder_dir=parent_dir+test_folder+'/'
    for prefix in prefixes:
        model_list=['model_ligand']
        if 'cnn' in prefix:
            model_list=['model_cnn_1','model_cnn_2','model_cnn_3','model_cnn_4','model_cnn_5']
        for model in model_list:
            for feat in num_feats:
                pharmacophore_file=folder_dir+prefix+str(feat)+'_'+model+'_predicted_pharmacophores.json'
                output=subprocess.check_output('python getf1.py '+pharmacophore_file+ ' ' +folder_dir+'pharmit_ligand_db' +' --actives ' + folder_dir +'actives_final.ism --decoys '+folder_dir+'decoys_final.ism' ,shell=True)
                output=output.decode('utf-8')
                results=output.split()
                f1=float(results[1])
                recall=float(results[3])
                precision=float(results[5])
                hit_rate=float(results[8])
                enrichment=float(results[11])
                df=df.append({'system':test_folder,'num_feats':feat,'model':model,'precision':precision,'recall':recall,'f1':f1,'enrichment':enrichment,'hit_rate':hit_rate},ignore_index=True)
print(df.head())
df.to_csv('dude_prediction_results.csv')
```

Obtain mean and max pharmacophore from ligand scores

```python
for test_folder in test_folders:
    target_df=pickle.load(open(parent_dir+test_folder+'/target_df.pkl','rb'))
    print(test_folder)
    print(np.max(target_df['f1']))
    target_df=target_df[target_df['file'].apply(lambda x: len(x.split('_'))>2)]
    target_df=target_df[target_df['file'].apply(lambda x: int(x.split('_')[1])<=5)]
    print(np.max(target_df['f1']),np.mean(target_df['f1']))
```
